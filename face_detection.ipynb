```python
import face_recognition
import cv2
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import os
import base64
from io import BytesIO
```

```python
# Image input

new_img = Image.open('/content/drive/MyDrive/python_ML/face/face.jpg')

# im_flip = ImageOps.flip(im)
width, height = new_img.size
resized = new_img.resize((width//3, height//3))

# # 현재사진이 뒤집혀있어서, 뒤집은 코드
# resized_flip = ImageOps.flip(resized)

# plt.imshow(resized) # Resized img check

# save resized img
resized_save = resized.save('resized.jpg' )
im = cv2.imread('/content/resized.jpg')
# print(im.shape)
```

```python


# Load the jpg file into a numpy array
image = face_recognition.load_image_file('/content/resized.jpg')


face_locations = face_recognition.face_locations(image, number_of_times_to_upsample=3, model="cnn")


list = []
for face_location in face_locations:
    # Print the location of each face in this image
    top, right, bottom, left = face_location
    face_image = image[top:bottom, left:right]
   
    pil_image = Image.fromarray(face_image)
    # pil_image.show(face_image)
    # plt.imshow(pil_image)
    pil_image.save('indi.jpg')

    list.append(pil_image)



 
    # print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

# print(list)

list_test = []

data = {}



for i in range(len(list)):
    buffered = BytesIO()
    list[i].save(buffered, format="JPEG")
    img_str = base64.b64encode(buffered.getvalue())
    list_test.append(img_str)

    data[f'image{i+1}'] = img_str.decode('utf-8')

print(data)

    
    # You can access the actual face itself like this:
  
```
